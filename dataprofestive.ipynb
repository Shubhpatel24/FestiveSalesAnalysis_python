#Exploratory Data Analysis

Gender

```python
sdata.columns
```

```python
ax = sns.countplot(x='Gender',data=sdata)

for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sdata.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
sales_gen = sdata.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x='Gender', y='Amount', data = sales_gen)
```

from above we can see that most of the buyers are females

Age

```python
sdata.columns
```

```python
ax = sns.countplot(data = sdata, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = sdata.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

from above graph womer byuers are more in the 25 to 35 age group

State

```python
sdata.columns
```

```python
sales_state = sdata.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(8)
sns.set(rc={'figure.figsize':(15,8)})
sns.barplot(data = sales_state, x = 'State',y= 'Orders')
```

```python
sales_state = sdata.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(8)
sns.set(rc={'figure.figsize':(15,8)})
sns.barplot(data = sales_state, x = 'State',y= 'Amount')
```

Marital status

```python
ax = sns.countplot(data = sdata, x = 'Marital_Status')


for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = sdata.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount', hue='Gender')
```

Ocupation

```python
ax = sns.countplot(data = sdata, x = 'Occupation')
sns.set(rc={'figure.figsize':(20,14)})
for bars in ax.containers:
    ax.bar_label(bars)
```

Product Category

```python
ax = sns.countplot(data = sdata, x = 'Product_Category')
sns.set(rc={'figure.figsize':(15,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = sdata.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(6)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount')
```

```python
sales_state = sdata.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(6)

sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders')
```
